```python
import os 
import torch
```

```python
grad_files = os.listdir("grads/")
grad_files = [(int(f[f.find("t_")+2:f.find("pt")-1]), f) for f in grad_files]
grad_files = sorted(grad_files, key=lambda x:x[0])
grad_files[:10]
    
```

```python
grads = [torch.load("grads/"+grad_file) for i, grad_file in grad_files]
```

```python
grads = [grad.abs().mean().item() for grad in grads]
```

```python
import matplotlib.pyplot as plt 
plt.plot(grads)
```

```python
plt.figure(figsize=(20,10))
plt.plot(grads)
```

```python
lagging = [(grads[i+1]-grads[i]).abs().mean() for i in range(len(grads)-1)]
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,10))
plt.plot(lagging)
```
